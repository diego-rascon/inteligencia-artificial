```python
import numpy as np

# Datos de entrada: características de las propiedades
x = np.array([
    [2310, 2, 2, 20],
    [2333, 2, 2, 1.2],
    [2356, 3, 1.5, 33],
    [2379, 3, 2, 43],
    [2402, 2, 3, 53],
    [2425, 4, 2, 23],
    [2448, 2, 1.5, 99],
    [2471, 2, 2, 34],
    [2494, 3, 3, 23],
    [2517, 4, 4, 55],
    [2540, 2, 3, 22]
])

# Datos de salida: precios de las propiedades
y = np.array([142000, 144000, 151000, 150000, 139000,
             169000, 126000, 142900, 163000, 169000, 149000])

# Función para calcular la media y el rango (se cambió np.ptp por np.std)
def get_params(A):
    p1 = np.mean(A, axis=0)   # Media de cada característica
    p2 = np.std(A, axis=0)    # Desviación estándar de cada característica
    return p1, p2

# Función para normalizar las características
def escalar(A, promedios, desviaciones):
    return np.divide(A - promedios, desviaciones)

# Función para calcular el parámetro theta y b usando descenso de gradiente
def get_theta(x, y, alpha=0.01, iteraciones=5000):
    m, n = x.shape  # Número de muestras (m) y características (n)
    
    # Inicializar parámetros
    theta = np.zeros([n, 1])  # Vector theta de tamaño (n, 1)
    b = 0                     # Bias inicializado a 0
    
    # Ajustar el tamaño de y para que coincida con x
    y = y.reshape([m, 1])
    
    # Bucle de descenso de gradiente
    for _ in range(iteraciones):
        # Calcular las predicciones
        h = np.dot(x, theta) + b
        
        # Calcular el error (h - y)
        error = h - y
        
        # Gradiente para theta y b
        gradient = (1/m) * np.dot(x.T, error)  # Gradiente para theta
        b_gradient = (1/m) * np.sum(error)     # Gradiente para el bias
        
        # Actualizar los parámetros
        theta -= alpha * gradient  # Actualizar theta
        b -= alpha * b_gradient    # Actualizar bias
    
    return b, theta

# Calcular la media y desviación estándar de x
promedios, desviaciones = get_params(x)

# Normalizar las características de x
x_normalizado = escalar(x, promedios, desviaciones)

# Ajustar los parámetros theta y b usando la versión corregida de get_theta
b, theta = get_theta(x_normalizado, y)

# Mostrar el valor de b y theta calculados
print(f'b = {b:.2f},\nw =\n{theta}')

# Definir un nuevo valor de propiedad y normalizarlo para la predicción
new_x = np.array([3333, 3, 3, 33])
new_x_normalizado = (new_x - promedios) / desviaciones

# Predicción para el nuevo valor usando los parámetros ajustados
precio_predicho = np.dot(new_x_normalizado, theta) + b
print(f'Precio predicho para el nuevo valor: {precio_predicho[0]:.2f}')
```
