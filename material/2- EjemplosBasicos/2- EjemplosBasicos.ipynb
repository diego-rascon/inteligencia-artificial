```python
import numpy as np

A = np.array([[1, 1, 3], [1, 1, 5], [0, 1, 1]])
# A = np.array([[1,1,0,3], [1,1,0,5], [0,1,0,1]])

print(f"A= \n{A}")
# invA = np.linalg.inv(A)
# Calcular la inversa de la matriz A
try:
    invA = np.linalg.inv(A)
except:
    print("La matriz no tiene inversa")
else:
    print(f"La inversa de A es= \n{invA}")
    print(f"El producto punto de A y su inversa es la matriz identidad= \n {
          A.dot(invA)}")

ren, col = A.shape
print(f"A tiene {ren} renglones y {col} columnas")

# Calcular el determinante de A
if ren != col:
    print("La matriz tiene que ser cuadrada para calcular el determinante")
else:
    detA = np.linalg.det(A)
    print(f"El determinante de A es= {detA}")

# Calcular la transpuesta
print(f"La transpuesta de A es=\n{np.transpose(A)}")
```


# Gráfica 3D
Aquí tienes la gráfica 3D de la función $𝑧 = sin(𝑥^2+𝑦^2)$

Utilizando matplotlib, hemos creado una malla de puntos en el rango de -5 a 5 tanto para 𝑥 como para 𝑦, y luego hemos graficado la superficie correspondiente. 

La barra de color a la derecha de la gráfica proporciona una referencia visual para los valores de z.

```python
# %matplotlib notebook
# %matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función matemática (por ejemplo, una función z = f(x, y))


def func(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


# Crear la malla de puntos (x, y)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = func(x, y)

# Crear la figura
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Añadir una barra de color para dar contexto a la altura
fig.colorbar(surf)

# Mostrar la gráfica
plt.show()
```
