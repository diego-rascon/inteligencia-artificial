```python
import numpy as np
import matplotlib.pyplot as plt
# import pandas as pd


A = np.array([[1, 1], [1, 2], [1, 3], [1, 4], [1, 5], [1, 6]])
Y = np.array([[3100], [4500], [4400], [5400], [7500], [8100]])

datos = np.hstack((A[:, 1:], Y))  # Horizontalmente apilado

# df = pd.DataFrame(datos)
# df.to_csv("resultados.csv", index=False, sep=',')

print('A= ', A, sep='\n', end='\n\n')
print('Y= ', Y, sep='\n', end='\n\n')
X = np.linalg.inv(A.T.dot(A)).dot(A.T).dot(Y)
print('X= ', X, sep='\n', end='\n\n')

fig1 = plt.figure()
plt.plot(A[:, 1], Y, 'r-')
plt.ylabel('salida sensor')
plt.xlabel('nivel real')
plt.title('Ecuacion Lineal')
Salida = A.dot(X)
plt.plot(A[:, 1], Salida, 'b-')
A = np.array([[1, 9]])
Salida = A.dot(X)
plt.plot(A[:, 1], Salida, 'g*')
print(f"si A={A}, entonces Y={Salida}")
```
