```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
marks = ['o', 's', 'v', '*', '.', 'd', 'D', '+']

center_1 = np.array([[1, 1]])
center_2 = np.array([[9, 9]])

points = np.random.random([20, 2]) * 10

distance_to_center_1 = np.linalg.norm(points - center_1, axis=1)
distance_to_center_2 = np.linalg.norm(points - center_2, axis=1)

colors = np.where(distance_to_center_1 < distance_to_center_2, 'y', 'm')

yellow_points = []
purple_points = []

for i in range(len(colors)):
    if (colors[i] == 'y'):
        yellow_points.append(points[i])
    else:
        purple_points.append(points[i])

yellow_average = np.mean(yellow_points, axis=0)
purple_average = np.mean(purple_points, axis=0)

center_size = 100
plt.grid(True)

plt.scatter(points[:, 0], points[:, 1], c=colors)
plt.scatter(center_1[0, 0], center_1[0, 1], c='y',
            s=center_size, marker='o', label='Centro 1')
plt.scatter(center_2[0, 0], center_2[0, 1], c='m',
            s=center_size, marker='o', label='Centro 2')

plt.scatter(yellow_average[0], yellow_average[1], c='y',
            s=center_size, marker='*', label='Yellow Average')
plt.scatter(purple_average[0], purple_average[1], c='m',
            s=center_size, marker='*', label='Purple Average')

plt.show()
```
